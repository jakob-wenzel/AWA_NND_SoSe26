# Bereinigen von jährliches Einkommen nach Städten je Einwohner 

```python
import pandas as pd

# Datei laden
file_path = "./data/raw/vgrdl_r2b3_bs2024.xlsx"

# Tabellenblatt "1.4" laden
df = pd.read_excel(file_path, sheet_name="1.4")

# Nur gewünschte Spalten auswählen
# A=0, B=1, C=2, H=7, AK=36
df_selected = df.iloc[:, [0, 1, 2, 7, 36]]

# Optional: Spalten umbenennen (sehr empfehlenswert!)
df_selected.columns = ["col_A", "col_B", "col_C", "col_H", "col_AK"]

# Als CSV speichern
df_selected.to_csv("./data/processed/cleaned_income.csv", index=False)

print(df_selected.head())
```

# Bereinigen von EV Anteil von Großstädten

```python
import pandas as pd

df = pd.read_csv("./data/raw/FZ Pkw mit Elektroantrieb Gemeinde_-4851084469163656342.csv")

# Zeilen filtern
df_filtered = df[
    df["Stadttyp"].notna() &
    (df["Stadttyp"] != "")
]

# Spalten auswählen
df_result = df_filtered[
    [
        "OBJECTID",
        "AGS",
        "Gemeinde",
        "Berichtszeitpunkt",
        "Gemeindeschlüssel",
        "Stadttyp",
        "Pkw Elektro Anteil"
    ]
]

df_result.to_csv("./data/processed/cleaned_EV_Anteil.csv", index=False)
```

# Bereinigen Ladeinfrastruktur

```python
import pandas as pd

df = pd.read_csv("./data/raw/Laderegister.csv", encoding="cp1252", skiprows=10, sep=";")

# Spalten auswählen
df_result = df[
    [
        "Postleitzahl",
        "Ort",
        "Anzahl Ladepunkte"
    ]
]

df_result.to_csv("./data/processed/cleaned_Laderegister.csv", index=False)
```
